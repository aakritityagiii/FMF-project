```python
import numpy as np
import pandas as pd
from pyDOE2 import lhs

# Settings
n_samples = 5000

# Parameter ranges
bounds = {
    'r1': [2, 6],
    'r2': [7, 12],
    'r3': [11, 16],
    'D1': [0.005, 0.03],
    'D2': [0.005, 0.03],
    'D3': [0.003, 0.02]
}

# Latin Hypercube Sampling
lhs_samples = lhs(6, samples=n_samples, criterion='maximin')

# Scale to actual ranges
r1 = bounds['r1'][0] + lhs_samples[:, 0] * (bounds['r1'][1] - bounds['r1'][0])
r2 = bounds['r2'][0] + lhs_samples[:, 1] * (bounds['r2'][1] - bounds['r2'][0])
r3 = bounds['r3'][0] + lhs_samples[:, 2] * (bounds['r3'][1] - bounds['r3'][0])
D1 = bounds['D1'][0] + lhs_samples[:, 3] * (bounds['D1'][1] - bounds['D1'][0])
D2 = bounds['D2'][0] + lhs_samples[:, 4] * (bounds['D2'][1] - bounds['D2'][0])
D3 = bounds['D3'][0] + lhs_samples[:, 5] * (bounds['D3'][1] - bounds['D3'][0])

# Dataframe
df = pd.DataFrame({
    'r1': r1,
    'r2': r2,
    'r3': r3,
    'D1': D1,
    'D2': D2,
    'D3': D3
})

# Constraint r1 < r2 < r3
df = df[df['r1'] < df['r2']]
df = df[df['r2'] < df['r3']]
df = df.reset_index(drop=True)

print(f"Total valid samples: {len(df)}")
print(df.head())
```

```python
df.to_csv('merged_file_5000.csv', index=False)
print("Saved! Total rows:", len(df))
```
