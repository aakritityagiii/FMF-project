```python
import numpy as np
import matplotlib.pyplot as plt

# Load data
true_A = np.load('true_inverse_A_FFNN.npy')
pred_A = np.load('pred_inverse_A_FFNN.npy')

true_B = np.load('true_inverse_B_FFNN.npy')
pred_B = np.load('pred_inverse_B_FFNN.npy')

params = ['r1','r2','r3','D1','D2','D3']

fig, axes = plt.subplots(2, 3, figsize=(12,8))

for i, param in enumerate(params):
    ax = axes[i//3, i%3]

    # Dataset A
    ax.scatter(true_A[:, i], pred_A[:, i], alpha=0.4, s=10, label='Dataset A')
    
    # Dataset B
    ax.scatter(true_B[:, i], pred_B[:, i], alpha=0.4, s=10, label='Dataset B')

    # Ideal line
    min_val = min(true_A[:, i].min(), true_B[:, i].min())
    max_val = max(true_A[:, i].max(), true_B[:, i].max())
    ax.plot([min_val, max_val], [min_val, max_val], 'k--')

    ax.set_title(f'FFNN Prediction — {param}')
    ax.set_xlabel(f'True {param}')
    ax.set_ylabel(f'Predicted {param}')

    if i == 0:
        ax.legend()

plt.tight_layout()
plt.savefig('FFNN_overlay.png', dpi=300, bbox_inches='tight')
plt.show()
```
